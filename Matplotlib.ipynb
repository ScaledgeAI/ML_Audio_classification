```python
x = np.arange(1,11)
y = 2*x+5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show
```

```python
##Compute the x and y coordinates for points on a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("Sinewave Form")

#Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#COmpute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'b--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('Cosine')

plt.show()
```

```python
# Bar plot
x = [2,8,10]
y = [11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title("Bar Graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")
```

```python
#Histograms
a = np.array([82,87,5,43,21,56,40,77,34,56])
plt.hist(a,bins=20)
plt.title("Histogram")
plt.show()
```

```python
#Box plot using Matplotlib
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist = True);
```

```python
#Pie Chart
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','cyan','red']
explode = (0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct = '%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
